# Sequence Modification Basics

Welcome to your first Python exercises! In these activities, you will modify working code to produce different output. Each exercise shows you code that works, and your task is to change it to meet the new requirements.

## What you'll practice
- Printing text with `print()`
- Simple arithmetic
- String concatenation (joining text together)

## How to work
1. Read the task description for each exercise
2. Look at the working code provided
3. Modify the code in the tagged cell to produce the expected output
4. The output shows you exactly what should appear when you run the code

**Remember**: At this stage, we're just using `print()` statements — no variables yet!
